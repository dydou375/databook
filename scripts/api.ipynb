```python
import requests
import time
import random

# Liste d'ISBN à tester (à compléter selon tes besoins)
isbns = [
    "9782226239938",  # Le Nom du vent
    "9780553381689"   # Le Meurtre de Roger Ackroyd
]

def get_livre_data(isbn):
    url = f"https://openlibrary.org/api/books?bibkeys=ISBN:{isbn}&format=json&jscmd=data"
    response = requests.get(url)
    data = response.json()
    key = f"ISBN:{isbn}"
    if key in data:
        livre = data[key]
        # Extraction des champs selon ton schéma SQL
        titre = livre.get('title')
        date_publication = livre.get('publish_date')
        langue = livre.get('languages', [{}])[0].get('key', '').split('/')[-1] if livre.get('languages') else None
        description = livre.get('notes') or livre.get('description', {}).get('value') if isinstance(livre.get('description'), dict) else livre.get('description')
        nb_pages = livre.get('number_of_pages')
        couverture_url = livre.get('cover', {}).get('large') or livre.get('cover', {}).get('medium')
        # Note moyenne non disponible sur Open Library, à compléter via d'autres sources
        note_moyenne = None
        # Auteurs
        auteurs = [a['name'] for a in livre.get('authors', [])]
        # Genres (subjects)
        genres = [s['name'] for s in livre.get('subjects', [])] if 'subjects' in livre else []
        return {
            "isbn": isbn,
            "titre": titre,
            "date_publication": date_publication,
            "langue": langue,
            "description": description,
            "note_moyenne": note_moyenne,
            "nb_pages": nb_pages,
            "couverture_url": couverture_url,
            "auteurs": auteurs,
            "genres": genres
        }
    else:
        print(f"ISBN {isbn} non trouvé.")
        return None

# Simulation d'une activité humaine
for isbn in isbns:
    livre_data = get_livre_data(isbn)
    if livre_data:
        print("Livre récupéré :")
        print(livre_data)
        print("-" * 40)
    # Pause aléatoire entre 2 et 5 secondes
    time.sleep(random.uniform(2, 5))
```

```python
import requests
import time
import random
import json
import os
from urllib.parse import quote_plus

categories = [
    {"fr": "Fiction", "en": "Fiction"},
    {"fr": "Science", "en": "Science"},
    # ... (toutes tes catégories ici)
]

max_results = 10  # Tu peux augmenter ce nombre si tu veux plus de livres par catégorie

# Crée un dossier pour stocker les fichiers JSON
os.makedirs("livres_json", exist_ok=True)

for cat in categories:
    sujet = cat["en"]  # On ne prend que la catégorie anglaise
    livres_json = []
    print(f"=== Livres pour le sujet : {sujet} (en) ===")
    url = (
        f"https://www.googleapis.com/books/v1/volumes"
        f"?q=subject:{quote_plus(sujet)}&maxResults={max_results}"
    )
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        livres = data.get("items", [])
        for livre in livres:
            info = livre.get("volumeInfo", {})
            titre = info.get("title", "Titre inconnu")
            auteurs = info.get("authors", ["Auteur inconnu"])
            langue_livre = info.get("language", "non précisée")
            notes = info.get("averageRating", None)
            nbr_vote = info.get("ratingsCount", None)
            resume = info.get("description", "description non présente")
            
            # Récupération des ISBN
            identifiers = info.get("industryIdentifiers", [])
            isbn_10 = next((id["identifier"] for id in identifiers if id["type"] == "ISBN_10"), None)
            isbn_13 = next((id["identifier"] for id in identifiers if id["type"] == "ISBN_13"), None)
            isbn = isbn_13 or isbn_10 or "N/A"

            # Récupération des edition_key via Open Library
            edition_keys = []
            if isbn != "N/A":
                ol_url = f"https://openlibrary.org/api/books?bibkeys=ISBN:{isbn}&format=json&jscmd=data"
                ol_res = requests.get(ol_url)
                if ol_res.status_code == 200:
                    ol_data = ol_res.json()
                    key = f"ISBN:{isbn}"
                    if key in ol_data:
                        edition_keys = ol_data[key].get("identifiers", {}).get("openlibrary", [])
            elif titre != "Titre inconnu" and auteurs[0] != "Auteur inconnu":
                q_title = quote_plus(titre)
                q_author = quote_plus(auteurs[0])
                ol_url = f"https://openlibrary.org/search.json?title={q_title}&author={q_author}"
                ol_res = requests.get(ol_url)
                if ol_res.status_code == 200:
                    ol_data = ol_res.json()
                    docs = ol_data.get("docs", [])
                    if docs:
                        edition_keys = docs[0].get("edition_key", [])

            # Ajout à la liste
            livres_json.append({
                "titre": titre,
                "auteurs": auteurs,
                "langue": langue_livre,
                "resume": resume,
                "isbn": isbn,
                "edition_keys": edition_keys,
                "note": notes,
                "nombre_votes": nbr_vote
            })

    # Sauvegarde dans un fichier JSON
    nom_fichier = f"livres_json/{sujet.replace(' ', '_').lower()}.json"
    with open(nom_fichier, "w", encoding="utf-8") as f:
        json.dump(livres_json, f, ensure_ascii=False, indent=2)
    print(f"Fichier créé : {nom_fichier} ({len(livres_json)} livres)\n")
    time.sleep(random.uniform(2, 5))
```
